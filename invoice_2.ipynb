```python
# Data Manipulation
import pandas as pd

# Data Visualisation
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
# Checking missing values
df.isna().sum()
df['LABOR_AMOUNT'] = df['LABOR_AMOUNT'].fillna(0)
```

```python
# Generating descriptive statistics summary
df.describe().round()
```

```python
# Getting month year from INVOICE_DATE
df['month_year'] = df['INVOICE_DATE'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
# grouping month_year by LABOR_AMOUNt
df_temp = df.groupby('month_year').sum()['LABOR_AMOUNT'].reset_index()
```

```python
# Setting the figure size
plt.figure(figsize=(16, 5))
plt.plot(df_temp['month_year'], df_temp['LABOR_AMOUNT'], color='#b80045')
plt.xticks(rotation='vertical', size=8)
plt.show()
```

```python
## This is the calculation of total labour amount w.r.t the market id
labor_amt = pd.DataFrame(df.groupby('MARKET_ID').sum()['LABOR_AMOUNT'])


labor_amt.sort_values(by=['LABOR_AMOUNT'], inplace=True, ascending=False)

labor_amt
```

```python
## This is the total labour amount w.r.t the market id

plt.figure(figsize=(10,6))
plt.bar('MARKET_ID','LABOR_AMOUNT',data=df,color='y')
plt.show()
```
